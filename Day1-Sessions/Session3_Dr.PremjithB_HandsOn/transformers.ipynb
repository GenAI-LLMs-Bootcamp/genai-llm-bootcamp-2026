transformers is Hugging Face's flagship library, giving you access to hundreds of thousands of pretrained models for NLP, vision, audio, and multimodal tasks — all behind a unified API.

The three core concepts:

* pipeline() — the simplest entry point. Pass a task name and it handles tokenization, inference, and decoding for you in one call.

* AutoTokenizer / AutoModel — for more control. Load any model checkpoint by name and work with it directly, useful for fine-tuning or custom inference logic.

* The Hugging Face Hub — where models live. from_pretrained("bert-base-uncased") downloads and caches weights automatically.

## Embedding

### Context
* River context:
The fisherman sat on the bank of the river, watching the water flow gently past.

* Finance context:
She went to the bank to deposit her salary.

* No contex: The word "bank" in isolation — as a traditional word2vec embedding would encode it.

```python
from transformers import AutoTokenizer, AutoModel
import torch
import torch.nn.functional as F

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model     = AutoModel.from_pretrained("bert-base-uncased", attn_implementation="eager") # Added attn_implementation='eager'
model.eval()

def get_bank_attention(sentence):
    inputs   = tokenizer(sentence, return_tensors="pt")
    tokens   = tokenizer.convert_ids_to_tokens(inputs["input_ids"][0])
    bank_idx = tokens.index("bank")

    with torch.no_grad():
        out = model(**inputs, output_attentions=True)

    # Shape: [12 layers][1, 12 heads, seq_len, seq_len]
    # We take the last layer, average across all 12 heads
    last_layer = out.attentions[-1]
    attn_heads = last_layer[0, :, bank_idx, :]
    avg_attn   = attn_heads.mean(dim=0)

    # Sort tokens by attention weight descending
    pairs = sorted(zip(tokens, avg_attn.tolist()),
                    key=lambda x: -x[1])

    print(f"\nSentence: {sentence}")
    print(f"'bank' is at token index {bank_idx}\n")
    print(f"{'Token':<14} {'Attn Weight':>12}  {'Bar'}")
    print("-" * 50)
    for tok, w in pairs:
        bar = "█" * int(w * 80)
        print(f"{tok:<14} {w:>12.4f}  {bar}")
    return dict(pairs)

s_river   = "The fisherman sat on the bank of the river, watching the water flow gently past."
s_finance = "She went to the bank to deposit her salary and apply for a loan."

attn_r = get_bank_attention(s_river)
attn_f = get_bank_attention(s_finance)

def get_embedding(sentence):
    inputs   = tokenizer(sentence, return_tensors="pt")
    tokens   = tokenizer.convert_ids_to_tokens(inputs["input_ids"][0])
    bank_idx = tokens.index("bank")
    with torch.no_grad():
        out = model(**inputs)
    return out.last_hidden_state[0, bank_idx, :]

emb_r = get_embedding(s_river)
emb_f = get_embedding(s_finance)

sim = F.cosine_similarity(emb_r.unsqueeze(0), emb_f.unsqueeze(0))
print(f"\nCosine similarity (river vs finance): {sim.item():.4f}")
```
